## Ejercicio 1
Somos una empresa que tiene varias tiendas de electrodomesticos. Necesitamos un programa para manejar las tiendas, ver las ventas que han tenido, cuántos empleados hay, etc... Para ello vamos a modelizar la tienda en una clase, que tendrá las siguientes características:
* Nombre clase: "Tienda"
* Atributos comunes:
    * Tipo: "Electrodomésticos"
    * Abierta: True
* Atributos propios de cada tienda:
    * Nombre - String
    * Dirección - String
    * Número de empleados - int
    * Ventas ultimos 3 meses - Lista de 3 numeros
* Método para calcular las ventas de todos los meses, que devuelva un numero con todas las ventas.
* Método que calcula la media de ventas de los ultimos meses, por empleado
* Método que devuelve en un string el nombre de la tienda, junto con su dirección.
* Método que obtiene las ventas del último mes.
* Método para dar una proyección de las ventas en caso de haber invertido X dinero en marketing en esa tienda. Siendo X el parámetro de entrada. Si X es menor de 1000, las ventas de los ultimos 3 meses hubiesen sido de (1.2 x ventas), si es mayor o igual de 1000, las ventas hubiesen sido de (1.5 x venta). El método reescribe el atributo de ventas en los últimos 3 meses, y además devuelve las ventas con el aumento.

Se pide:
1. Implementar la clase "Tienda"
2. Crear tres tiendas con datos inventados
3. Comprueba en al menos una de ellas, todo lo implementado en la clase tienda (sus atributos, media de ventas, ventas/empleado...)
3. Calcular las ventas del último mes de todas las tiendas. Para ello usa el bucle `for`
4. Imprime por pantalla los nombres de las tiendas cuya dirección lleve el string "Avenida"
5. Seguro que ya has hecho este apartado... Documenta la clase :)

```python
# DEFINIR CLASE TIENDA

class Tienda:
    '''
    La clase Tienda es para definir diferentes tiendas con sus caracteristicas
    '''
    tipo = 'Electrodomesticos'
    abierta = True

    def __init__(self, nombre, direccion, num_empleados, ventas_ultimos_3_meses):
        '''
        Describe como es la tienda
        '''
        self.nombre = nombre
        self.direccion = direccion
        self.num_empleados = num_empleados
        self.ventas_ultimos_3_meses = ventas_ultimos_3_meses
        
    def datos_tienda(self):
        '''
        Mostrar las caracteristicas de cada tienda
        '''
        out = 'Nombre: ' + self.nombre + '\nDireccion: ' + self.direccion + '\nNº empleados: ' + str(self.num_empleados) + '\nVentas en los ultimos 3 meses: ' + str(self.ventas_ultimos_3_meses)
        return out
    
    def ventas(self):
        '''
        Calcular las ventas de todos los meses
        '''
        return sum(self.ventas_ultimos_3_meses)
    
    def ventas_empleado(self):
        '''
        Calcular la media de ventas por empleado
        ''' 
        return self.ventas() / self.num_empleados
    
    def nombre_tienda(self):
        '''
        Devuelve el nombre de la tienda y su direccion
        ''' 
        return self.nombre + ' ' + self.direccion
    
    def ventas_ultimo_mes (self):
        '''
        Calcular las ventas del ultimo mes
        ''' 
        return self.ventas_ultimos_3_meses[-1]
    
    def proyeccion_ventas (self, marketing):
        '''
        Calcular la proyección de las ventas en caso de haber invertido X dinero en marketing en esa tienda
        '''
        if marketing < 1000:
            return (1.2 * self.ventas()) + self.ventas()
        
        elif marketing >= 1000:
            return (1.5 * self.ventas()) + self.ventas()
```

```python
# 2. Crear tres tiendas con datos inventados
tienda_1 = Tienda(nombre = 'Primera tienda',
                 direccion = 'Avenida España, 1',
                 num_empleados = 10,
                 ventas_ultimos_3_meses = [1000, 2000, 1500])

tienda_2 = Tienda(nombre = 'Segunda tienda',
                 direccion = 'Calle Madrid, 2',
                 num_empleados = 5,
                 ventas_ultimos_3_meses = [700, 750, 650])

tienda_3 = Tienda(nombre = 'Tercera tienda',
                 direccion = 'Avenida Las Rozas, 3',
                 num_empleados = 20,
                 ventas_ultimos_3_meses = [5000, 6000, 4000])

# 3. Comprueba en al menos una de ellas, todo lo implementado en la clase tienda (sus atributos, media de ventas, ventas/empleado...) 
print(tienda_1.datos_tienda())
print(tienda_1.ventas())
print(tienda_1.ventas_empleado())
print(tienda_1.nombre_tienda())
print(tienda_1.ventas_ultimo_mes())
print(tienda_1.proyeccion_ventas(500))

# 4. Calcular las ventas del último mes de todas las tiendas. Para ello usa el bucle for
lista_tiendas = [tienda_1.ventas_ultimo_mes(), tienda_2.ventas_ultimo_mes(), tienda_3.ventas_ultimo_mes()]

ventas_totales_ultimo_mes = 0

for i in lista_tiendas:
    
    ventas_totales_ultimo_mes += i

print('Las ventas totales del ultimo mes son de: ', ventas_totales_ultimo_mes)

# 5. Imprime por pantalla los nombres de las tiendas cuya dirección lleve el string "Avenida"
```
