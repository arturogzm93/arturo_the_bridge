```python
import sqlite3
import pandas as pd
dir_db = "chinook.db"
conn = sqlite3.connect(dir_db)
cursor = conn.cursor()
query = """
SELECT *
FROM albums
"""
mi_query = cursor.execute(query)

mi_query = cursor.execute("SELECT name FROM sqlite_master WHERE type ='table';")

mi_query.fetchall()

pd.read_sql_query(query, conn)
```

```python
query = """
SELECT * 
FROM customers
JOIN invoices ON customers.CustomerId = invoices.CustomerId
"""

pd.read_sql_query(query, conn)
```

```python
# 1
query = """
SELECT c.LastName 
FROM customers c
JOIN invoices i ON c.CustomerId = i.CustomerId
"""

pd.read_sql_query(query, conn)
```

```python
# 2
query = """
SELECT c.FirstName, i.InvoiceId, i.InvoiceDate, i.BillingCountry 
FROM customers c
JOIN invoices i ON c.CustomerId = i.CustomerId
WHERE c.Country = 'Brazil'
"""

pd.read_sql_query(query, conn)
```

```python
# 3
query = """
SELECT e.FirstName || ' ' || e.LastName as 'Nombre empleado', i.*
FROM employees e
JOIN customers c ON e.EmployeeId = c.SupportRepId
JOIN invoices i ON c.CustomerId = i.CustomerId
"""
mi_df = pd.read_sql_query(query, conn)
mi_df
```

```python
# 4
query = """
SELECT c.FirstName || ' ' || c.LastName as 'Nombre empleado', c.Country, e.LastName, i.Total
FROM employees e
JOIN customers c ON e.EmployeeId = c.SupportRepId
JOIN invoices i ON c.CustomerId = i.CustomerId
"""
mi_df = pd.read_sql_query(query, conn)
mi_df
```

```python
# 5
query = """
SELECT t.Name as 'Nombre de la cancion', it.InvoiceId
FROM invoice_items it
JOIN tracks t ON it.TrackId = t.TrackId
"""
mi_df = pd.read_sql_query(query, conn)
mi_df
```

```python
# 6
query = """
SELECT t.Name, t.Bytes, AlbumId, GenreId 
FROM tracks t
"""
mi_df = pd.read_sql_query(query, conn)
mi_df
```

```python
# 7
query = """
SELECT *
FROM playlists p
JOIN playlist_tracks pt ON p.PlaylistId = pt.PlaylistId
"""
mi_df = pd.read_sql_query(query, conn)
mi_df
```
