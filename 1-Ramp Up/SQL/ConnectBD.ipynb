```python
import sqlite3 #driver para sqlite
import pandas as pd
```

```python
dirDb="chinook.db"#nombre de la bd
con = sqlite3.connect(dirDb)#conesión a la bd
```

```python
# se crea el cursor
cursor = con.cursor()
```

```python
q = """Select * 
from genres;""" #query
```

```python
tabla = cursor.execute(q)
```

```python
tabla.fetchall()
```

```python
consulta ="""Select * 
from tracks inner join genres on genres.GenreId=tracks.GenreId
where genres.GenreId = 4;"""
```

```python
tabla = cursor.execute(consulta)
```

```python
tabla.fetchall()
```

```python
brasilClientes = """select distinct customers.CustomerId, customers.FirstName,customers.LastName, invoices.InvoiceId,
invoices.InvoiceDate,invoices.BillingCountry
from customers inner join invoices on customers.CustomerId = invoices.CustomerId
where customers.country = 'Brazil';"""
tabla = cursor.execute(brasilClientes)
pd.read_sql_query(brasilClientes, con)
```

```python
brasilClientes = """select invoices.*
from customers inner join invoices on customers.CustomerId = invoices.CustomerId
where customers.country = 'Brazil';"""
tabla = cursor.execute(brasilClientes)
pd.read_sql_query(brasilClientes, con)
```

```python
#4
query = """select c.FirstName || ' ' || c.LastName as nombre, c.Country, e.LastName, i.Total
from employees e inner join customers c on e.EmployeeId = c.SupportRepId
inner join invoices i on c.CustomerId=i.CustomerId;"""
tabla = cursor.execute(query)
pd.read_sql_query(query, con)
```

```python
#5
query = """select i.InvoiceId,t.Name
from invoice_items i inner join tracks t on i.TrackId=t.TrackId;"""
tabla = cursor.execute(query)
pd.read_sql_query(query, con)
```

```python
#6
query = """Select t.Name,m.Name as format, a.Title, g.Name
from tracks t inner join genres g on t.GenreId=g.GenreId
inner join media_types m on t.MediaTypeId=m.MediaTypeId
inner join albums a on t.AlbumId=a.AlbumId;"""
tabla = cursor.execute(query)
pd.read_sql_query(query, con)
```

```python
#7
query = """Select p.Name, count(pt.TrackId)
from playlists p inner join playlist_track pt on p.PlaylistId=pt.PlaylistId
group by p.Name;"""
tabla = cursor.execute(query)
pd.read_sql_query(query, con)
```

```python
#8
query = """Select e.EmployeeId,e.FirstName,e.LastName, sum(i.Total)
from employees e inner join customers c on e.EmployeeId=c.CustomerId
inner join invoices i
group by e.EmployeeId, e.FirstName,e.LastName;"""
tabla = cursor.execute(query)
pd.read_sql_query(query, con)
```

```python
#9
query = """Select e.EmployeeId,e.FirstName,e.LastName, sum(i.Total) as ventas
from employees e inner join customers c on e.EmployeeId=c.CustomerId
inner join invoices i
where i.InvoiceDate like '2009%'
group by e.EmployeeId, e.FirstName,e.LastName"""
tabla = cursor.execute(query)
pd.read_sql_query(query, con)
```

```python
#10
query = """Select a.ArtistId, sum (i.Total)
from artists a inner join albums al
inner join tracks t on al.AlbumId=t.AlbumId
inner join invoice_items it on t.TrackId=it.TrackId
inner join invoices i on it.InvoiceId=i.InvoiceId
group by a.ArtistId;"""
tabla = cursor.execute(query)
pd.read_sql_query(query, con)
```
